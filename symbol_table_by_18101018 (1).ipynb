Implement the following functionalities:
1. Insert a new symbol/name along with its type into the symbol table
2. Search a symbol/name along with its type from the symbol table
3. Delete a symbol/name along with its type into the symbol table
4. Show the contents of symbol table in the console
5. Update an already existing entry in the symbol table
6. Get Hash Key

```python
import pandas as pd
import random
symbol_table = pd.DataFrame(columns=['Hash','Name','Type'])
```

```python
def insert(symbol_table, name, type):
  hash_key = getHashKey(name)
  if (hash_key not in symbol_table.Hash.values):
    symbol_table.loc[hash_key] = [hash_key,name,type]
    print(symbol_table.to_string(index=False))
  elif (hash_key in symbol_table.Hash.values) and (name not in symbol_table.Name.values):
    symbol_table.loc[random.randint(hash_key,hash_key+20)] = [hash_key,name,type]
    print(symbol_table.to_string(index=False))
  else:
    print("The value is already in the symbol table")

def search(symbol_table, name):
  if (name in symbol_table.Name.values):  
    print(symbol_table.loc[symbol_table['Name'] == name].to_string(index=False))
  else:
    print("The value is not found in the symbol table")

def delete(symbol_table,name):
  if (name in symbol_table.Name.values):
    index = symbol_table.loc[symbol_table['Name'] == name].index[0]
    symbol_table.drop([index],axis=0,inplace=True)  
    print(symbol_table.to_string(index=False))
  else:
    print("The value is not found in the symbol table")

def update(symbol_table,name,updated_name,updated_type):
  if (name in symbol_table.Name.values):
    index = symbol_table.loc[symbol_table.Name == name,'Name'].index[0]
    symbol_table.loc[index,['Name']] = updated_name
    symbol_table.loc[index,['Type']] = updated_type
    print(symbol_table.to_string(index=False))
  else:
    print("The value is not found in the symbol table")

def getHashKey(name):
    ascii_value = 0
    try:
      for char in name:
        ascii_value += ord(char) #ord - will convert each character in the key to the ASCII value 
      return ascii_value 
    except:
      ascii_value=name

def showHashKey(name):
    ascii_value = 0
    try:
      for char in name:
        ascii_value += ord(char) #ord - will convert each character in the key to the ASCII value 
      print(ascii_value)
    except:
      ascii_value=name
      print(ascii_value)
    

```

```python
while True:
    print("Press 1 for insert")
    print("Press 2 for search")
    print("Press 3 for delete")
    print("Press 4 for show")
    print("Press 5 for update")
    print("Press 6 for getHashKey")
    print("Press 7 for exit")
    
    user_choice = int(input('Enter your choice: '))

    if user_choice == 1:
      n = input('Enter name: ')
      t = input('Enter type: ')
      insert(symbol_table, n, t)
    if user_choice == 2:
      n=input('Enter name: ')
      search(symbol_table, n)
    if user_choice == 3:
      n=input('Enter name: ')
      delete(symbol_table, n)
    if user_choice == 4:
      print(symbol_table.to_string(index=False))
    if user_choice == 5:
      en=input('Enter exiting name: ')
      un=input('Enter updated name: ')
      ut=input('Enter updated type: ')
      update(symbol_table, en, un, ut)
    if user_choice == 6:
      n=input('Enter name: ')
      showHashKey(n)
    if user_choice == 7:
      break
```
